## seção 1 , trabalhando com Pandas SERIES


uma serie em pandas é como uma coluna em um Tabela, uma matriz de apenas 1 coluna 

```python
### utilizando o comando pd.series

idade = pd.Series([20,50,40,33,80])

print(idade)
```

```python
### retornar apena os valores da Series.

idade.values
```

```python
### retornar apenas os indices dos registros

idade.index
```

```python
#### alterando indices...

indice_novo = ['Rafaela','Eder','Dandara','Icaro', 'Beethoven']

idade_list = [34,35,1,12,2]

nova_serie = pd.Series(idade_list,index= indice_novo)

print(nova_serie)
```

Acessando informações de uma Seriess

```python
print(nova_serie['Rafaela'])
```

### Estatisticas Basicas ###

```python
df.describe()
```

### Estatisticas de Dados, categorias ( não numericos )

para estatisticas de variaveis precisamos adicionar o argumento include =(0)

TOP: O é o valor que aparece com mais aparece 

FREQ:  quantidade de vezes que o valor mais comum aparece QUE O VALOR COMUM APARECE(TOP)

COUNT: (Quantidade de elemento que aparece existente em campos )

UNIQUE: Quantidade de elementos que aparece no campos UNICOS sem repetição 

```python
df.describe(include=('O'))
```

## Estatisticas Basicas de todos os Dados

ainda possivel aplicar o pd.decribe a todos os dados, Para isso, precisamos adicionar o argumento include ='all'

```python
df.describe(include=('all'))
```
